```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Média de idade = 38 anos, renda anual = 57.407, valor da compra = 425.63, pontuação de fidelidade 6.7, frenquência de compra = 19

---

### Quantidade de compras por região

```python
df["region"].value_counts()
```

---

### Fidelidade baseado em pontos abaixo da média 5

```python
compra = df[df["loyalty_score"] <= 5.0]
compra.value_counts()
```

---

### Pessoas com renda anual alta usando 70000 como ponto de partida para análise


```python
renda_anual = df[df["annual_income"] >= 70000]
renda_anual.value_counts()
```

---

### Renda anual de acordo com a idade, percebe-se que pessoas com idade acima de 50 anos possui uma renda de 70000 acima.


```python

idade = plt.figure(figsize=(9,3))
renda_anual= idade.add_axes([0,0,1,1])
renda_anual.bar(df["age"], df["annual_income"], facecolor ="#00DB33")
plt.grid(True)
plt.xlabel("Idade")
plt.ylabel("Renda anual")
plt.title("Renda anual baseado pela idade")
plt.show()

```

---

### Com essa análise percebe-se que pessoas com idade acima de 50 anos compram mais, em média 25 compras por ano.


```python

compra_idade = plt.figure(figsize=(9,3))
quant_compra= compra_idade.add_axes([0,0,1,1])
quant_compra.bar(df["age"], df["purchase_frequency"], facecolor ="#DB5E0A")
plt.grid(True)
plt.xlabel("Idade")
plt.ylabel("Quantidade de compra")
plt.title("Quantidade de compra baseado pela idade")
plt.show()
```

---

### De acordo com os dados, pessoas com maior renda anual moram no Sul e leste do país.


```python

renda_anual = plt.figure(figsize=(9,3))
regiao = renda_anual.add_axes([0,0,1,1])
regiao.bar(df["region"], df["annual_income"], facecolor ="#DC2616")
plt.grid(True)
plt.xlabel("Região")
plt.ylabel("Renda anual")
plt.title("Região com maior renda anual")
plt.show()
```

---

### Pessoas com idade entre 38 a 42 tende a comprar mais.


```python

idade1 = plt.figure(figsize=(9,3))
age1 = idade1.add_axes([0,0,1,1])
age1.hist(df["age"], facecolor ="#4404DE", edgecolor="white")
plt.grid(True)
plt.title("Distribuiçõa por idade")
plt.xlabel("Idade")
plt.show()
```
